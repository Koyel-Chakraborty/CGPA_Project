```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('cgpa.xlsx')
```

# Subsetting and labelling the data

## Subsetting Data

```python
df_M = df[df['Gender']=='M']
```

```python
df_F = df[df['Gender']=='F']
```

## Labelling Data

### Change Row Label 

```python
df.set_index('Name',inplace = True)
df_M.set_index('Name',inplace = True)
df_F.set_index('Name',inplace = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
import statsmodels.api as sm
```

```python
#Combined Group
plt.rcParams['font.family']='serif'
plt.rcParams['font.serif']='Times New Roman'
fig,ax = plt.subplots(2,2,figsize = (10,12),gridspec_kw = {'hspace':0.5,'wspace':0.4})
fig.suptitle('\nCGPA Scores',fontsize = 20)

ax[0][0].hist(df['CGPA'],color = 'red',edgecolor = 'blue', alpha = 0.5)
for label in (ax[0][0].get_xticklabels() + ax[0][0].get_yticklabels()):
    label.set_fontsize(16)
ax[0][0].set_title('Histogram\n',fontsize = 18)
ax[0][0].set_xlabel(' CGPA',fontsize = 16)
ax[0][0].set_ylabel('Frequency',fontsize = 16)

sns.kdeplot(ax = ax[0][1],data = df['CGPA'], color = 'green',alpha = 0.5, shade = True)
for label in (ax[0][1].get_xticklabels() + ax[0][1].get_yticklabels()):
    label.set_fontsize(16)
ax[0][1].set_title('KDE Plot\n',fontsize = 18)
ax[0][1].set_xlabel(' CGPA',fontsize = 16)
ax[0][1].set_ylabel('Density',fontsize = 16)

sns.boxplot(ax = ax[1][0],data = df['CGPA'], color = 'blue', showmeans = True, meanprops = {'markerfacecolor':'red'})
for label in (ax[1][0].get_xticklabels() + ax[1][0].get_yticklabels()):
    label.set_fontsize(16)
for patch in ax[1][0].artists:
    r, g, b, a = patch.get_facecolor()
    patch.set_facecolor((r, g, b, 0.3))
ax[1][0].set_xticklabels(' ',fontsize = 18)
ax[1][0].set_title('Box Plot\n',fontsize = 18)
ax[1][0].set_ylabel(' CGPA',fontsize = 16)

sm.qqplot(df['CGPA'], color = '#8e4585',ax = ax[1][1], fit = True, line = 's',alpha = 0.5)
ax[1][1].set_xlim(-3,3)
ax[1][1].set_ylim(-3,3)
for label in (ax[1][1].get_xticklabels() + ax[1][1].get_yticklabels()):
    label.set_fontsize(16)
ax[1][1].set_title('Quantile-Quantile Plot\n',fontsize = 18)
ax[1][1].set_xlabel('Theoritical Quantiles',fontsize = 16)
ax[1][1].set_ylabel('Sample Quantiles',fontsize = 16)

plt.savefig("Combined.png", dpi = 400)
```

```python
#Male-Female Group
plt.rcParams['font.family']='serif'
plt.rcParams['font.serif']='Times New Roman'
fig,ax = plt.subplots(2,2,figsize = (10,12),gridspec_kw = {'hspace':0.5,'wspace':0.4})

ax[0][0].hist(df_M['CGPA'],color = '#48AAAD',edgecolor = 'blue', alpha = 0.5)
for label in (ax[0][0].get_xticklabels() + ax[0][0].get_yticklabels()):
    label.set_fontsize(16)
ax[0][0].set_title('Histogram\n(Male Group)',fontsize = 18)
ax[0][0].set_xlabel(' CGPA',fontsize = 16)
ax[0][0].set_ylabel('Frequency',fontsize = 16)

ax[0][1].hist(df_F['CGPA'],color = '#f25278',edgecolor = 'red', alpha = 0.5)
for label in (ax[0][1].get_xticklabels() + ax[0][1].get_yticklabels()):
    label.set_fontsize(16)
ax[0][1].yaxis.set_ticks(range(0,4,1))
ax[0][1].set_title('Histogram\n(Female Group)',fontsize = 18)
ax[0][1].set_xlabel(' CGPA',fontsize = 16)
ax[0][1].set_ylabel('Frequency',fontsize = 16)

sns.kdeplot(ax = ax[1][0],data = df_M['CGPA'], color = '#48AAAD',alpha = 0.5, shade = True)
for label in (ax[1][0].get_xticklabels() + ax[1][0].get_yticklabels()):
    label.set_fontsize(16)
ax[1][0].set_title('KDE Plot\n(Male Group)',fontsize = 18)
ax[1][0].set_xlabel(' CGPA',fontsize = 16)
ax[1][0].set_ylabel('Density',fontsize = 16)

sns.kdeplot(ax = ax[1][1],data = df_F['CGPA'], color = '#f25278',alpha = 0.5, shade = True)
for label in (ax[1][1].get_xticklabels() + ax[1][1].get_yticklabels()):
    label.set_fontsize(16)
ax[1][1].set_title('KDE Plot\n(Female Group)',fontsize = 18)
ax[1][1].set_xlabel(' CGPA',fontsize = 16)
ax[1][1].set_ylabel('Density',fontsize = 16)

plt.savefig("gender_eda_1.png", dpi = 400)
```

```python
#Male-Female Group
plt.rcParams['font.family']='serif'
plt.rcParams['font.serif']='Times New Roman'
fig,ax = plt.subplots(2,2,figsize = (10,12),gridspec_kw = {'hspace':0.5,'wspace':0.4})

sns.boxplot(ax = ax[0][0],data = df_M, y = 'CGPA', color = '#48AAAD', showmeans = True, meanprops = {'markerfacecolor':'blue'})
for label in (ax[0][0].get_xticklabels() + ax[0][0].get_yticklabels()):
    label.set_fontsize(16)
for patch in ax[0][0].artists:
    r, g, b, a = patch.get_facecolor()
    patch.set_facecolor((r, g, b, 0.5))
ax[0][0].set_ylim(5,9)
ax[0][0].set_xticklabels(' ',fontsize = 18)
ax[0][0].set_title('Box Plot\n(Male Group)',fontsize = 18)
ax[0][0].set_ylabel(' CGPA',fontsize = 16)

sns.boxplot(ax = ax[0][1],data = df_F['CGPA'], color = '#f25278', showmeans = True, meanprops = {'markerfacecolor':'red'})
for label in (ax[0][1].get_xticklabels() + ax[0][1].get_yticklabels()):
    label.set_fontsize(16)
for patch in ax[0][1].artists:
    r, g, b, a = patch.get_facecolor()
    patch.set_facecolor((r, g, b, 0.5))
ax[0][1].set_ylim(5,9)
ax[0][1].set_xticklabels(' ',fontsize = 18)
ax[0][1].set_title('Box Plot\n(Female Group)',fontsize = 18)
ax[0][1].set_ylabel(' CGPA',fontsize = 16)

sm.qqplot(df_M['CGPA'], color = '#48AAAD',ax = ax[1][0], fit = True, line = 's',alpha = 0.5)
ax[1][0].set_xlim(-3,3)
ax[1][0].set_ylim(-3,3)
for label in (ax[1][0].get_xticklabels() + ax[1][0].get_yticklabels()):
    label.set_fontsize(16)
ax[1][0].set_title('Quantile-Quantile Plot\n(Male Group)',fontsize = 18)
ax[1][0].set_xlabel('Theoritical Quantiles',fontsize = 16)
ax[1][0].set_ylabel('Sample Quantiles',fontsize = 16)
ax[1][0].get_lines()[1].set_color('blue')

sm.qqplot(df_F['CGPA'], color = '#f25278',ax = ax[1][1], fit = True, line = 's',alpha = 0.5)
ax[1][1].set_xlim(-3,3)
ax[1][1].set_ylim(-3,3)
for label in (ax[1][1].get_xticklabels() + ax[1][1].get_yticklabels()):
    label.set_fontsize(16)
ax[1][1].set_title('Quantile-Quantile Plot\n(Female Group)',fontsize = 18)
ax[1][1].set_xlabel('Theoritical Quantiles',fontsize = 16)
ax[1][1].set_ylabel('Sample Quantiles',fontsize = 16)


plt.savefig("gender_eda_2.png", dpi = 400)
```
